# <b>File Handling 😹</b>
### File handling pada Python adalah proses dalam membuat, membaca, mengubah, dan menghapus file pada sistem. Python menyediakan fungsi utama dalam file handling, yaitu ```open()```. Fungsi ```open()``` mengambil dua parameter, yaitu ```filename``` dan ```mode```.

---

## Mode dalam fungsi ```open()``` yang sering digunakan sebagai berikut:
### 1. 'r' => <b>(Read) </b> berfungsi untuk membuka dan membaca sebuah file.
### 2. 'a' => <b>(Append)</b> berfungsi untuk membuka file untuk ditambahkan sebuah data, jika belum ada maka Python akan membuat secara otomatis.
### 3. 'w' => <b>(Write)</b> berfungsi untuk membuka file untuk ditulis sebuah data, jika tidak ada maka akan dibuat secara otomatis.


### Setelah membuka file dengan fungsi ```open()```, kita perlu menutupnya dengan fungsi ```close()```. Tujuannya adalah untuk menjaga resource memori dan mencegah kebocoran memori.


---

# Menyimpan data dengan database sederhana menggunakan CSV

### CSV (Comma Separated Value) adalah sebuah file teks yang menyimpan data dalam format tabel terstruktur dengan memisahkan nilai-nilai menggunakan koma atau titik koma.

## Mode 'r' (Read)

````python
file = open("data_mahasiswa.csv", "r")
data = file.read()
print(data)
file.close() # tutup file setelah proses membuka file selesai
````

## Mode 'a' (Append)

````python
file = open('data_mahasiswa.csv', 'a')
# menambahkan data ke file csv dengan fungsi write
file.write('Nama, NIM, Jurusan\n')
file.write('Fauzul, 232410101015, Sistem Informasi\n')
file.write('Eka, 232410102004, Teknologi Informasi\n')
file.write('Taufiq, 232410103025, Informatika\n')
file.close() # tutup file setelah proses menambah data selesai

# membaca file setelah ditambahkan data menggunakan mode 'r'
file = open('data_mahasiswa.csv', 'r')
data = file.read() # membaca file setelah ditambahkan data
print(data)
file.close() # tutup file setelah proses membaca file selesai
````

## Mode 'w' (Write)

````python
file = open('data_mahasiswa.csv', 'w')
file.write('Nama : Budiono Siregar\nNIM : 242410104100\nJurusan : Teknik Komputer dan Jaringan\n')

file = open ('data_mahasiswa.csv', 'a')
file.write ('Nama saya ganti Taufik\nNIM saya segitu\nJurusan Surabaya Malang\n')
file.close()

# uncomment kode berikut & lihat perbedaannya
# file = open('mahasiswa.txt', 'w') #
# file.write('Nama : Fauzul Akbar\nNIM : 242410101001\nJurusan : Sistem Informasi\n')
# file.close()

file = open('data_mahasiswa.csv', 'r')
data = file.read()
print(data)
file.close()
````

### Setiap menggunakan fungsi ```open()``` untuk membuka file, kita juga harus menggunakan fungsi ```close()``` untuk menutup file.
### Cara lain agar dalam melakukan File Handling lebih efisien adalah dengan menggunakan statement ```with```. Statement ini memungkinkan kita melakukan File Handling secara efisien, karena setelah kita membuka file maka otomatis akan ditutup jika sudah selesai.

````python
# Contoh penggunaan with statement
with open('data_mahasiswa.csv', mode='r') as file:
    data = file.read()
    print(data)
````

## Penerapan statement 'with' dalam File Handling

````python
# Menulis file dengan with statement menggunakan mode 'w' (write)
with open('data_mahasiswa.csv', 'w') as file:
    file.write('Nama, NIM, Jurusan\n')

# Menambahkan data ke file dengan with statement menggunakan mode 'a' (append)
with open('data_mahasiswa.csv', 'a') as file:
    file.write('Rasthian, 222410102000, Teknologi Informasi\n')
    file.write('Fathan, 232410103015, Informatika')

# Membaca file dengan with statement menggunakan mode 'r' (read)
with open('data_mahasiswa.csv', 'r') as file:
    data = file.read()
    print(data)
````

# Merging data

### Menggunakan `join()`

````python
data1 = {
    'Nama' : ['Kambing', 'Sapi', 'Kerbau'],
    'Umur' : [3, 4, 5]
}

data2 = {
    'Alamat' : ['Jember', 'Manchester United']
}

data1 = pd.DataFrame(data1)
data2 = pd.DataFrame(data2)

dataJoin = data1.join(data2)
dataJoin
````

### Menggunakan `concat()`

````python
data1 = {
    'Nama' : ['Kambing', 'Sapi', 'Kerbau'],
    'Umur' : [3, 4, 5]
}

data2 = {
    'Alamat' : ['Jember', 'Manchester United', 'Jl. Kalimantan 9']
}

data1 = pd.DataFrame(data1)
data2 = pd.DataFrame(data2)

dataJoin = pd.concat([data1, data2], axis=0)
dataJoin
````

### Menggunakan `merge()`

````python
data1 = {
    'Nama' : ['Kambing', 'Sapi', 'Kerbau'],
    'Umur' : [3, 4, 5]
}

data2 = {
    'Alamat' : ['Jember', 'Manchester United', 'Jl. Kalimantan 9']
}

data1 = pd.DataFrame(data1)
data2 = pd.DataFrame(data2)

dataMerge = pd.merge(data1, data2, left_index=True, right_index=True)
dataMerge
````

````python
data1 = pd.DataFrame({
    'Nama': ['Kambing', 'Sapi', 'Kerbau'],
    'Umur': [3, 4, 5]
})

data2 = pd.DataFrame({
    'Nama': ['Kambing', 'Sapi', 'Kerbau'],
    'Alamat': ['Jember', 'Manchester United', 'Jl. Kalimantan 9']
})

# Menggabungkan berdasarkan kolom 'Nama'
dataMerge = pd.merge(data1, data2, on='Nama')
print(dataMerge)
````
